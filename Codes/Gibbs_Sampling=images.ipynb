# Text Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
STOPWORDS = set(stopwords.words('indonesian'))
from nltk.stem import WordNetLemmatizer


def cleansing(df):
    # List of words to remove
    words_to_remove = ['bola', 'detik', 'okezone', 'kompas', 'cnbcindonesia']
    df_clean = df.str.lower() #to lowercase
    df_clean = [re.sub(r'^.*?-\s*', '', i) for i in df_clean] # remove all the introduction related about media (for ex: kota diterbitkan, dll), in news
    df_clean = [re.sub(r'.com', '', i) for i in df_clean] # remove .com
    df_clean = [re.sub(r'.tv', '', i) for i in df_clean] # remove .tv for kompas
    df_clean = [re.sub(r'#\S+', '', i) for i in df_clean] # remove hashtags
    df_clean = [re.sub(r'https?:\/\/.*[\r\n]*', '', i, flags=re.MULTILINE) for i in df_clean] # remove https
    df_clean = [re.sub(r"\d+", "", i) for i in df_clean] #remove (digit)
    df_clean = [re.sub(r'[^\w]', ' ', i) for i in df_clean] # remove not word
    df_clean = [re.sub(r'\s+', ' ', i) for i in df_clean] # remove space > 1

    # Remove specific words
    df_clean = [' '.join(word for word in text.split() if word not in words_to_remove) for text in df_clean]

    # Apply stopwords removal to each row in the DataFrame
    df_clean = [' '.join(word for word in text.split() if word not in STOPWORDS) for text in df_clean]


    return df_clean
```

Setelah menjalankan fungsi ini, teks akan menjadi lowercase, tanpa angka, hashtag, kata-kata “https”,
ataupun “.com”, hanya berisi karakter word, tidak ada spasi berlebih, dan tidak ada kata-kata STOPWORDS.  

Selain requirement cleaning pada soal, disini saya sengaja melakukan cleaning pada nama dari media pada clean_text dikarenakan jika nama media tidak dibersihkan dari teks, model klasifikasi berita mungkin akan mempelajari bias dari media tersebut. Misalnya, semua berita dari “cnbcc” dalam dataset saya berlabel “politik”, model mungkin akan belajar bahwa setiap berita yang menyebutkan “cnbcc” kemungkinan besar berhubungan dengan politik. # Text Representation & Modelling

Requirement:

**size vektor adalah 50, dan frequency kata yang diperhitungkan minimal 3**

Berdasarkan Requirement tersebut maka pendekatan yang dilakukan adalah word embedding, dan saya memilih untuk menggunakan teknik Word2Vec(Skip-gram) dan FastText.

Teknik Skip-gram dan FastText dapat memenuhi persyaratan ukuran vektor 50 dan frekuensi minimal 3 dengan alasan berikut:
1. **Skip-gram (Word2Vec)**: Dalam model Word2Vec, ukuran vektor menentukan dimensi ruang vektor di mana kata-kata akan dipetakan. Dengan ukuran vektor 50, setiap kata akan direpresentasikan sebagai vektor 50 dimensi dalam ruang tersebut. Parameter min_count dalam Word2Vec dapat digunakan untuk mengabaikan semua kata dengan frekuensi total lebih rendah dari ini. Jadi, dengan mengatur min_count menjadi 3, kata-kata yang muncul kurang dari 3 kali di seluruh dokumen akan diabaikan. Saya memilih skipgram disini dikarenakan:
  - Perform better dengan Kata yang Jarang Muncul: Skip-gram cenderung lebih baik daripada CBOW dengan kata-kata yang jarang muncul. Karena artikel berita sering mengandung terminologi spesifik atau entitas bernama yang tidak sering muncul, Skip-gram bisa lebih cocok.

  - Menangkap Lebih Banyak Konteks: Skip-gram belajar untuk memprediksi konteks yang diberikan kata, yang bisa bermanfaat ketika konteks membawa informasi penting untuk mengklasifikasikan artikel berita.

2. **FastText**: FastText memenuhi persyaratan ukuran vektor 50 dan frekuensi kata minimal 3 karena menggunakan pendekatan subword embedding dan parameter min_count. Setiap kata dipecah menjadi subword yang lebih kecil. Kemudian, setiap subword direpresentasikan sebagai vektor 50 dimensi. Vector representation dari text diperoleh dengan menggabungkan vektor representation subword. Parameter min_count digunakan untuk mengabaikan kata-kata yang jarang muncul.

W2F (Skip-gram) architecture

```python
from gensim.models import Word2Vec
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np
import nltk


class SkipGram(BaseEstimator, TransformerMixin):
    def __init__(self, vector_size=50, window=5, min_count=3, workers=4):
        self.vector_size = vector_size
        self.window = window
        self.min_count = min_count
        self.workers = workers
        self.word2vec = None

# tokenizer
    def fit(self, X, y=None):
        sentences = [nltk.word_tokenize(text) for text in X]
        self.word2vec = Word2Vec(sentences, vector_size=self.vector_size, window=self.window, min_count=self.min_count, workers=self.workers, sg=1)

        return self

    def transform(self, X):
        return np.array([np.mean([self.word2vec.wv[w] for w in nltk.word_tokenize(text) if w in self.word2vec.wv]
                                 or [np.zeros(self.vector_size)], axis=0)
                         for text in X])
```

Mengubah teks menjadi vektor menggunakan model Word2Vec (Skip-gram) dari library Gensim dengan memberikan parameter value sesuai dengan requirements.

**Pipeline construction & Modelling**

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


pipeline_w2v_rf = Pipeline([
    ('w2v', SkipGram()),
    ('clf', RandomForestClassifier()),  # Random Forest classifier
])

pipeline_ft_rf = Pipeline([
    ('ft', Fasttext()),
    ('clf', RandomForestClassifier()),  # Random Forest classifier
])

pipeline_ft_svm = Pipeline([
    ('ft', Fasttext()),
    ('clf', SVC()),  # SVM classifier
])

pipeline_w2v_svm = Pipeline([
    ('w2v', SkipGram()),
    ('clf', SVC()),  # SVM classifier
])
```

Membuat 4 pipeline text representation, menggunakan Word2Vec skipgram dan FastText.

- **Pipeline Word2Vec**: Pipeline ini menggunakan Word2VecVectorizer untuk mengubah teks menjadi vektor dengan model Word2Vec (skipgram).

- **Pipeline FastText**: Pipeline ini menggunakan FastTextVectorizer untuk mengubah teks menjadi vektor dengan model FastText.

Feature Engineering



```python
from sklearn.pipeline import FeatureUnion


# Define the Word2Vec part of the pipeline
pipeline_w2v = Pipeline([
    ('w2v', SkipGram()),
])

# Define the FT part of the pipeline
pipeline_ft = Pipeline([
    ('ft', Fasttext()),
])

# Combine the FEATURE FT and Word2Vec parts using FeatureUnion
combined_features = FeatureUnion([
    ('ft', pipeline_ft),
    ('w2v', pipeline_w2v),
])


X_transformed = combined_features.fit_transform(x_train)

# Transform the test data
X_test_transformed = combined_features.transform(x_test)
```

Pemilihan parameter:

1. **'C': [0.1, 1, 10, 100]**: Parameter 'C' dalam SVM adalah parameter regularisasi yang membantu mengendalikan overfitting. Nilai yang lebih rendah menghasilkan batas keputusan yang lebih halus, sementara nilai yang lebih tinggi cenderung menghasilkan batas keputusan yang lebih kompleks untuk memasukkan lebih banyak titik data. Dalam hal ini, kita mencoba berbagai nilai, mulai dari 0.1 (lebih bias terhadap bias) hingga 100 (lebih bias terhadap varians).

2. **'kernel': ['linear', 'rbf', 'poly']**: Kernel adalah fungsi yang digunakan untuk mengubah ruang input menjadi ruang berdimensi tinggi. Pilihan kernel ini mencakup linear (untuk data yang linearly separable), rbf (Radial basis function, untuk data yang non-linearly separable), dan poly (polynomial, untuk data yang dapat dipisahkan dengan batas keputusan polinomial).

3. **'gamma': ['scale', 'auto']**: Parameter 'gamma' digunakan dalam kernel non-linear dan menentukan sejauh mana pengaruh titik data tunggal. Pilihan 'scale' dan 'auto' memungkinkan model untuk secara otomatis menentukan nilai gamma berdasarkan statistik data.

4. **'degree': [3, 4, 5]**: Parameter 'degree' digunakan dalam kernel polinomial dan menentukan derajat polinomial. Dalam hal ini, kita mencoba polinomial derajat 3, 4, dan 5.

Dengan mencoba berbagai kombinasi parameter ini, kita dapat menemukan model yang paling efektif untuk data dan task.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import random
import pickle
import os
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')


class Preprocessing:

    def __init__(self):
        pass

    def Reduce_inflection(self,Corpus):
        lemmatizer = WordNetLemmatizer()

        lemmatized_data = []

        for doc in Corpus:
            doc_list = []
            for word in doc.split(' '):
                if len(word) <= 3 or any(map(str.isdigit, word)) == True:
                    continue
                lemma = lemmatizer.lemmatize(word)
                doc_list.append(lemma)
            lemmatized_data.append(' '.join(doc_list))

        return np.array(lemmatized_data)

    def _preprocess_(self,_data):

        data = self.Reduce_inflection(_data)

        # remove words which occur in more than 90% of the documents, remove stopwords and keep the rare words.
        doc_vectorizer = CountVectorizer(max_df = 0.9,stop_words='english')  # max_df=0.90, min_df=2, max_features = 10000
        tf_data = doc_vectorizer.fit_transform(data).toarray()
        self.tf_data = tf_data


        print(f'Vocabulary size = {len(doc_vectorizer.vocabulary_.keys())}')

        tokenized_data = []  # tf_data is in sparse matrix form. tokenized_data[] will only store the indices of words present, with repetations, for each doc in corpus.

        for doc_idx, doc in enumerate(tf_data):
            embedded_doc = []
            non_zero_words = np.where(doc > 0)[0].tolist()

            for word_idx,word in enumerate(non_zero_words):
                embedded_doc = embedded_doc + [word for _ in range(doc[word])]

            tokenized_data.append(embedded_doc)

        self.inv_mapping = { v: k for k, v in  doc_vectorizer.vocabulary_.items()}
        return tokenized_data, len(doc_vectorizer.vocabulary_.keys())


class LDA:

    def __init__(self) -> None:
        pass

    def _fit_(self,tokenized_data, n_words,n_iterations = 10, n_topics = 10):
        self.data = tokenized_data  # data
        self.n_topics = n_topics

        n_docs = len(tokenized_data) # number of documents

        # n_words = Vocabulary size
        # dirichlet parameters initialisation
        alpha = 0.1
        beta = 0.1

        # other initialisations

        phi_t_w = np.zeros((n_topics,n_words))  # per-topic word distributions
        theta_d_t = np.zeros((n_docs,n_topics)) # per-document topic distributions

        n_w_in_d = np.zeros((n_docs))  # number of tokens in documents --> self.data[i]
        n_t_in_c = np.zeros((n_topics)) # number of times the topics are present in the corpus --> self.data

        topic_d_w = [[0 for _ in range(len(doc))] for doc in self.data] # t_d_w[i,j] = topic of self.data[i,j], numpy initialisation not done due to presence of docs of varied length.

        for doc_indx,doc in enumerate(self.data):
            for w_indx, word in enumerate(doc):
                # assign a random topic to each word in the beginning
                topic = random.randint(0,n_topics-1)

                topic_d_w[doc_indx][w_indx] = topic

                theta_d_t[doc_indx, topic] += 1
                phi_t_w[topic ,word] += 1

                n_w_in_d[doc_indx] += 1
                n_t_in_c[topic] += 1

        # training phase
        for _ in range(n_iterations):
            print(f"Iteration : {_+1}")
            for doc_indx,doc in enumerate(self.data):
                for w_indx,word in enumerate(doc):

                    topic = topic_d_w[doc_indx][w_indx]

                    # remove topic and word from the existing distribution
                    n_t_in_c[topic] -= 1
                    n_w_in_d[doc_indx] -= 1
                    phi_t_w[topic,word] -= 1
                    theta_d_t[doc_indx,topic] -= 1

                    # sampling new topic from multinomial distribution

                    prob_d_t = (theta_d_t[doc_indx] + alpha)/(n_w_in_d[doc_indx] + alpha*n_topics)
                    prob_t_w = (phi_t_w[:,word] + beta) / (beta*n_words + n_t_in_c)

                    prob_topic = prob_d_t * prob_t_w

                    prob_topic = prob_topic / np.sum(prob_topic)


                    new_topic = np.random.multinomial(1, prob_topic).argmax() # sample from multinomial distribution

                    # updation after sampling
                    n_t_in_c[new_topic] += 1
                    n_w_in_d[doc_indx] += 1
                    phi_t_w[new_topic,word] += 1
                    theta_d_t[doc_indx,new_topic] += 1
                    topic_d_w[doc_indx][w_indx] = new_topic

        self.theta = theta_d_t
        self.phi = phi_t_w


    def plot_distribution(self,doc_indx):

        plt.plot(self.theta[doc_indx]/np.sum(self.theta[doc_indx]))
        plt.title(f"Distribution of topics in document {doc_indx}")
        plt.xlabel("Topics")
        plt.ylabel("Probability values")
        #plt.show()
        plt.savefig(f't_100_Document_{doc_indx}.png')
        plt.clf()
        #plt.imsave(f'Outputs/epoch{epoch}_step{count}.jpg', img, cmap='gray')

    def show_top_(self,n_topics,k,index_to_token):
        temp = 0
        for i in range(self.n_topics):
            temp += 1
            print(f"Topic number {i}:")
            topic = self.phi[i]
            sorted_indices = topic.argsort()[::-1][:k]
            print([index_to_token[x] for x in sorted_indices])
            if temp == n_topics:
                break





if __name__ == '__main__':

    Data = pd.read_csv('DataBase.csv')
    Abstracts = Data['Abstract'].values

    print(f'Number of documents = {Abstracts.shape[0]}')

    preprocess = Preprocessing()
    tokenized_data, n_tokens = preprocess._preprocess_(Abstracts)

    number_of_topics = 100

    lda = LDA()
    lda._fit_(tokenized_data,n_tokens,n_iterations = 100,n_topics = number_of_topics)

    for i in range(10):
        lda.plot_distribution(i)

    lda.show_top_(100,10,preprocess.inv_mapping)

    store_objects = {0:preprocess, 1:lda}

    path = './Store_1000.pkl'

    with open(path, 'wb') as f:
        pickle.dump(store_objects,f,pickle.HIGHEST_PROTOCOL)


```
